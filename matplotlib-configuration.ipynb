# Interfaces

* Different interfaces exist for matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

# First plot

```python
ax = plt.subplot(111)
x = np.linspace(0, 1, 200)
y = 2 * x + np.random.randn(len(x))*0.3
ax.plot(x, y, label='my label')
ax.set_title('My title')
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.legend()
```

# Change style

* Style settings can be modified directly from `plt.rcParams`
* I.e. change the width of the lines:
* The same can be done setting individually each line (``linewidth=5``)

```python
plt.rcParams['lines.linewidth'] = 5
ax = plt.subplot(111)
x = np.linspace(0, 1, 200)
y = 2 * x + np.random.randn(len(x))*0.3
ax.plot(x, y, label='my label')
ax.set_title('My title')
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.legend()
```

# Example

* The first example shows a plot using `mystyle`
* This is the base style which I use for my plot
* It can be combined with other styles for special purposes

```python
def plot_sin():
    ax = plt.subplot(111)
    L = 2*np.pi
    x = np.linspace(0, L)
    ncolors = len(plt.rcParams['axes.prop_cycle'])
    shift = np.linspace(0, L, ncolors, endpoint=False)
    for s in shift:
        ax.plot(x, np.sin(x + s))
    ax.margins(0)
    ax.set_title('Sinusoids')
    return ax

plt.rcParams.update(plt.rcParamsDefault) # reset to default style
plt.style.use(['mystyle', 'mystyle-nb'])
plot_sin()
```

* I use now different colors
* `vega` colorstyle 

```python
plt.rcParams.update(plt.rcParamsDefault) # reset to default style
plt.style.use(['mystyle', 'mystyle-nb', 'mystyle-vega'])
plot_sin()
```

* Add extra settings for paper quality plots
* Add minot ticks
* Use ``Times`` fonts

```python
plt.rcParams.update(plt.rcParamsDefault) # reset to default style
plt.style.use(['mystyle', 'mystyle-paper', 'mystyle-vega'])
plot_sin()
```

* Same plot monocrome with different line styles. 
* It is useful for printing in black and white
    * You don't want to pay journal fees for color plots 

```python
plt.rcParams.update(plt.rcParamsDefault) # reset to default style
plt.style.use(['mystyle', 'mystyle-monocrome', 'mystyle-paper'])
plot_sin()
```

* Increase the size of the labels and line thickness for presentation
* It is very annoying when you can't read labels in a presentation. Isn't it?

```python
plt.rcParams.update(plt.rcParamsDefault) # reset to default style
plt.style.use(['mystyle', 'mystyle-pres'])
plot_sin()
```

* Same plot using now `bmh` style

```python
plt.rcParams.update(plt.rcParamsDefault) # reset to default style
plt.style.use('bmh')
plot_sin()
```

# Examples

* The `Matplotlib` website has a large gallery of examples
* http://matplotlib.org/gallery.html
* It can be a source of inspiration

```python
plt.rcParams.update(plt.rcParamsDefault)
plt.style.use(['mystyle', 'mystyle-vega'])
x = np.linspace(0, 4*np.pi, 201)  # define x between 0 and 4*pi
y = np.sin(x)                     # define y = sin(x)
ax = plt.subplot(111)            # define an axes object (similar to MLab)
ax.plot(x, y, label='$sin(x)$', linestyle='solid', linewidth=3, marker='o', 
        markevery=10, markeredgecolor='none', markersize=10)
ax.plot(x, y+0.5, label='$sin(x)+0.5$', linestyle='dashed', color='purple', 
        linewidth=5)
ax.fill_between(x, -1, np.abs(y-0.5)-1, label='$|sin(x)-0.5|-1$', color='red', 
        linewidth=3, alpha=0.5)
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.legend(loc='upper left', bbox_to_anchor=(1, 1))
ax.set_xticks(np.linspace(0, 4*np.pi, 5))
ax.set_xticklabels(['$0$', '$\pi$', '$2\pi$', '$3\pi$', '$4\pi$'], size='small')
ax.set_yticklabels(['-1', '-0.5', '0', '0.5', '1', '1.5'], family='serif', 
                   size='small')
ax.set_xlim([0, 4*np.pi])
ax.set_title('Test Matplotlib', family='serif', size='medium')
ax.minorticks_on()
```
