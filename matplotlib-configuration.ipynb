# Matplotlib

* Standard `python` plot module
* Inspired from `matlab`

# Interfaces

* Different interfaces exist for matplotlib
* The most common used is `pyplot`
* Conventionally abbreviated in `plt`

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

# Style

* The default style settings can be visualized using the command:

```python
plt.rcParamsDefault
```

# Change style

* Style settings can be modified directly from `plt.rcParams`
* I.e. change the width of the lines:

```python
plt.rcParams['axes.linewidth'] = 2
```

# Customize plots with style sheets

* It is possible to define customized style sheets
* `matplotlib` has already some style sheets:

```python
plt.style.available
```

# Personal styles

* Personal style sheets can be defined
* Define a file `<style-name>.mplstyle` and save in the default `mpl_configdir` of your system
* Generally `$HOME/.config/matplotlib/stylelib`

```python
import matplotlib
import os
print(matplotlib.get_configdir())
os.listdir(os.path.join(matplotlib.get_configdir(), 'stylelib'))
```

* Different styles can be combined
* I.e. a basic style is defined, but special configuration can be used for **papers** or **presentations**

```python
plt.style.use(['mystyle', 'mystyle-nb'])
```

# Example

* The first example shows a plot using `mystyle`

```python
def plot_sin():
    ax = plt.subplot(111)
    L = 2*np.pi
    x = np.linspace(0, L)
    ncolors = len(plt.rcParams['axes.color_cycle'])
    shift = np.linspace(0, L, ncolors, endpoint=False)
    for s in shift:
        ax.plot(x, np.sin(x + s), '-')
    ax.margins(0)
    return ax

plot_sin()
```

* Same plot using now `ggplot` style

```python
plt.rcParams.update(plt.rcParamsDefault) # reset to default style
plt.style.use('ggplot')
plot_sin()
```

# Example plot

```python
plt.rcParams.update(plt.rcParamsDefault)
plt.style.use(['mystyle', 'mystyle-nb'])
x = np.linspace(0, 4*np.pi, 201)  # define x between 0 and 4*pi
y = np.sin(x)                     # define y = sin(x)
ax = plt.subplot(111)            # define an axes object (similar to MLab)
ax.plot(x, y, label='$sin(x)$', linestyle='solid', linewidth=3, marker='o', 
        markevery=10, markeredgecolor='none', markersize=10)
ax.plot(x, y+0.5, label='$sin(x)+0.5$', linestyle='dashed', color='purple', 
        linewidth=5)
ax.fill_between(x, -1, np.abs(y-0.5)-1, label='$|sin(x)-0.5|-1$', color='red', 
        linewidth=3, alpha=0.5)
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.legend(loc='best')
ax.set_xticks(np.linspace(0, 4*np.pi, 5))
ax.set_xticklabels(['$0$', '$\pi$', '$2\pi$', '$3\pi$', '$4\pi$'], size='small')
ax.set_yticklabels(['-1', '-0.5', '0', '0.5', '1', '1.5'], family='serif', 
                   size='small')
ax.set_xlim([0, 4*np.pi])
ax.set_title('Test Matplotlib', family='serif', size='medium')
ax.minorticks_on()
```
